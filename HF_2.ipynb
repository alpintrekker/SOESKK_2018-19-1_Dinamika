```python
%matplotlib inline
import numpy as np
from IPython.display import Latex
```

```python
w = 0.8  # rad/sec
w_ = Latex(r'$\omega =' + str(w) + r'\, [rad/s]$')
a_n = 26  # m/sec**2
a_nt= Latex(r'$a_{n} =' + str(a_n) + r'\, [m/s^2]$')
```

# Dinamika 2. házi feladat megoldása
## Készítette: Kovács András (N80PN7)
### A/2

Adott:  {{w_}}, {{a_nt}}

```python
T = 2 * np.pi / w
T_t = Latex(r'$$T =\frac{2\pi}{\omega}=' + str(np.round(T, 3)) + r'\, [s],$$')
```

Ebből az egyenletes körmozgás periódusideje: {{T_t}}



```python
R = a_n / w**2
R_ = Latex(r'$$ R = \frac{a_n}{\omega^2}=' + str(np.round(R,3)) + r'\, [m],$$')
```

sugara: {{R_}}

```python
v = (a_n * R)**0.5
v_ = Latex(r'$$ v = \sqrt{a_n \cdot R \,} =' + str(np.round(v,3)) + r'\, [m/s]$$')
```

sebessége: {{v_}}

```python
n = 60 / T
n_ = Latex(r'$$ n = \frac{60\,[s/min]}{T[s]} =' + str(np.round(n,3)) + r'\, [1/min].$$')
```

fordulatszáma: {{n_}}

```python
t = np.linspace(0,40,2000)
fi = w * t
```

A $\phi - t$ függvény:

```python
from matplotlib import pyplot as plt
from matplotlib import ticker
f1, ax1 = plt.subplots(figsize=(16,9))
ax1.plot(t, fi/np.pi)
ax1.set_ylabel('$\phi$ [rad]', fontsize=16)
ax1.set_xlabel('$t$ [s]', fontsize=16)
ax1.yaxis.set_major_formatter(ticker.FormatStrFormatter('%g $\pi$'))
ax1.yaxis.set_major_locator(ticker.MultipleLocator(base=1.0))
ax1.grid()
u = ax1.plot()
```

Az $\omega - t$ függvény:

```python
omega_plot = t * 0 + w
f2, ax2 = plt.subplots(figsize=(16,9))
ax2.plot(t, omega_plot)
ax2.set_ylabel('$\omega$ [rad/s]', fontsize=16)
ax2.set_xlabel('$t$ [s]', fontsize=16)
u = ax2.plot()
```

```python
T_b = 0.2
T_b_t = Latex(r'$ T =' + str(T_b) + r'\, [s]$')
v_max = 1.2
v_max_t = Latex(r'$ v_{max} =' + str(v_max) + r'\, [m/s]$')
tau = 0.06
tau_t = Latex(r'$ \tau =' + str(tau) + r'\, [s]$')
```

### B/2

Adottak a harmonikus rezgőmozgás következő paraméterei {{T_b_t}}, {{v_max_t}}, {{tau_t}}. Ebből a hiányzó jellemzők:

```python
nu = 1 / T_b
nu_t =  Latex(r'$$\nu =\frac{1}{T}=' + str(np.round(nu, 3)) + r'\, [Hz],$$')
```

frekvencia: {{nu_t}}

```python
w_b = 2 * np.pi / T_b
w_b_t =  Latex(r'$$\omega =\frac{2\pi}{T}=' + str(np.round(w_b, 3)) + r'\, [rad/s],$$')
```

szögsebesség: {{w_b_t}}

```python
A = v_max / w_b
A_t =  Latex(r'$$ A =\frac{v_{max}}{\omega}=' + str(np.round(A, 3)) + r'\, [m],$$')
```

amplitúdó: {{A_t}}

```python
a_max = A * w_b**2
a_max_t = Latex(r'$$ a_{max} =A \cdot \omega ^2 =' + str(np.round(a_max, 3)) + r'\, [m/s^2].$$')
```

a legnagyobb gyorsulás: {{a_max_t}}

```python
x_tau = A * np.sin(w_b * tau + 0.5 * np.pi)
x_tau_t = Latex(r'$$X(\tau) = A \cdot \sin(\omega\tau + \alpha) =' + str(np.round(x_tau,3)) + '\, [m] $$')
```

#### B/2/I
Az $X(t)=A\cdot\sin(\omega t + \alpha)$ egyenlettel leírt harmonikus rezgőmozgás esetén az $X(0)=A$ peremfeltételből következik, hogy $\alpha = \pi /2 $. Ekkor {{x_tau_t}} -nek adódik.

#### B/2/II

```python
t2 = np.linspace(0,0.4,2000)
X = A * np.sin(w_b *t2 + 0.5 * np.pi)
V = A * w_b * np.cos(w_b *t2 + 0.5 * np.pi)
a =  - A * w_b**2 * np.sin(w_b *t2 + 0.5 * np.pi)
f3, ax3 = plt.subplots(figsize=(16,9))
ax3.plot(t2, X)
ax3.plot(tau, x_tau, 'ro')
ax3.set_ylabel('$X(t)$ [m]', fontsize=16)
ax3.set_xlabel('$t$ [s]', fontsize=16)
ax3.grid()
```

#### B/2/III

```python
f4, ax4 = plt.subplots(figsize=(16,16))
ax4.plot(X, V, label=u'Sebesség')
ax4.set_ylabel('$v(t)\,[m/s]$', fontsize=16)
ax4.set_xlabel('$x\, [m]$', fontsize=16)
ax5 = ax4.twinx()
ax5.plot(X, a, color='r', label=u'Gyorsulás')
ax5.set_ylabel('$a(t)\, [m/s^2]$ ', fontsize=16)
lines4,labels4 = ax4.get_legend_handles_labels()
lines5,labels5 = ax5.get_legend_handles_labels()
l = ax5.legend(lines4 + lines5, labels4 + labels5,loc=0, fontsize=14)
```

#### B/2/IV
